## Create a probability range for Home Goals and Away Goals
In order to simulate games we will generate random numbers between 0 and 1. We want to equate those numbers to a number for goals scored based on the probabilty distribution we have calculated.

Eg. We estimate that there is a 4% chance Team A scores 0 goals and a 14% chance that Team A scores 1 goal. We want it so that if the random number generated is between 0 and 0.04, 0 goals are scored. If the number generated is between 0.04 and 0.18, 1 goal is scored and so on.

```python
def goal_probs(goals):
    for i in range(0,10):
        if i == 0:
            y=goals[i]
            #d = {i:[0,y]} #this works testing a quicker method
            d={i:y}
            #print(x)
        else:
            #z = np.array([y,y+test[i]])
            z = [y, y+goals[i]]
            #d1 = {i:[y,y+goals[i]]} this works testing a quicker method
            d1 ={i:y+goals[i]}
            d.update(d1)
            y =y + goals[i]
    d.update({10:1})
    return d
    
single_line["Prob_Dist_HG"] = single_line["Prob_HG"].apply(lambda x: goal_probs(x))
single_line["Prob_Dist_AG"] = single_line["Prob_AG"].apply(lambda x: goal_probs(x))
print(single_line.iloc[0,]["Prob_Dist_HG"])
single_line.head()
```

### Applying the function to simulate home goals and away goals
we'll stick to 1000 simulations to start off with. We may go with more eventually

```python
#generate random numbers between 0 and 1. each number simulates a game. 

# def siumulate_goals(goal_prob):
#     goals=[]
#     sim = np.random.random_sample(1000)
#     for i in sim:
#         x = max(np.where((pd.DataFrame(goal_prob) <i).max())[0])
#         goals.append(x)
#     return goals

def sim_goal2(goal_prob):
    #print(goal_prob)
    goals=[]
    sim = np.random.random_sample(1000)
    for i in sim:
        x = min({key: value for (key, value) in goal_prob.items() if i < value })
        goals.append(x)
    return goals

single_line["HGsims"] = single_line["Prob_Dist_HG"].apply(lambda x: sim_goal2(x))
single_line["AGsims"] = single_line["Prob_Dist_AG"].apply(lambda x: sim_goal2(x))
single_line.head()
```

### Generating results from simulated games
From the simulated Home and Away goals, we will calculate points scored for each team and the goal difference for each team.<br>
We need to calculate both Home Points and Away Points as well as Home Goals and Away Goals so that we can restructure the data to extract performance by team

```python
def points(teamA,teamB):
    gd = np.array(teamA)-np.array(teamB)
    points = np.where(gd>0,3,np.where(gd==0,1,0))
    return points

def goal_dif(teamA,teamB):
    gd = np.array(teamA)-np.array(teamB)
    return gd
    
single_line["homepoints"] = single_line.apply(lambda row: points(row["HGsims"],row["AGsims"]), axis = 1)
single_line["awaypoints"] = single_line.apply(lambda row: points(row["AGsims"],row["HGsims"]), axis = 1)
single_line["HomeGD"] = single_line.apply(lambda row: goal_dif(row["HGsims"],row["AGsims"]), axis = 1)
single_line["AwayGD"] = single_line.apply(lambda row: goal_dif(row["AGsims"],row["HGsims"]), axis = 1)
single_line.head(2)
```

## Turning simulated fixtures into Team points and goals for and against

```python
home_teams = single_line[["league","season","matchdate","fixture","season_gw","Team_Home","Team_Away",
                          "Pred_HG","Pred_AG","HGsims","AGsims","HomeGD","homepoints"]].copy()
away_teams = single_line[["league","season","matchdate","fixture","season_gw","Team_Home","Team_Away",
                          "Pred_HG","Pred_AG","HGsims","AGsims","AwayGD","awaypoints"]].copy()
home_teams["HomeGame"]=1
away_teams["HomeGame"]=0
home_teams.rename(columns ={"Team_Home":"Team"
                           ,"Team_Away":"Opponent"
                           ,"Pred_HG":"Pred_GF"
                            ,"Pred_AG":"Pred_GA"
                           ,"HGsims":"GoalsFor"
                           ,"AGsims":"GoalsAgainst"
                            ,"HomeGD":"GD"
                           ,"homepoints":"points"},inplace =True)
away_teams.rename(columns ={"Team_Away":"Team"
                           ,"Team_Home":"Opponent"
                           ,"Pred_AG":"Pred_GF"
                            ,"Pred_HG":"Pred_GA"
                           ,"AGsims":"GoalsFor"
                           ,"HGsims":"GoalsAgainst"
                            ,"AwayGD":"GD"
                           ,"awaypoints":"points"},inplace =True)

def results(points,point):
    result =[]
    for i in points:
        if i ==point:
            output = 1
        else:
            output = 0
        result.append(output)
    return np.array(result)
    
    

col_order = home_teams.columns
away_teams = away_teams[col_order]

team_sims = pd.concat([home_teams,away_teams])
team_sims["Wins"] = team_sims["points"].apply(lambda x: results(x,3))
team_sims["Draws"] = team_sims["points"].apply(lambda x: results(x,1))
team_sims["Losses"] = team_sims["points"].apply(lambda x: results(x,0))

team_sims.sort_values(["Team","matchdate"]).head(3)
    
```

Write a function that ranks every season simulation per team. <br>
We'll use the zipped which contains each seasons points total per team. We'll use the index of the team in the summary table to extract the relevant points total per team. <br>
We'll also allocate a position group for each finishing position

```python
def rank_array(ind):
    ranks = []
    final = []
    for i in zipped:
        #print(i)
        rk = pd.Series(i).rank(method = "min", ascending = False)
        place = rk[ind]
        if place ==1:
            pos = "Winner"
        elif place <=4: 
            pos = "Champions League"
        elif place <=6:
            pos = "Europa League"
        elif place <=10:
            pos = "Upper Half"
        elif place <=17:
            pos = "Lower Half"
        else:
            pos = "Relegated"
        #print(rk[0])
        #print(pd.Series(i).rank(method = "min",ascending = False))# can print this as an array for ease of visualising
        ranks.append(rk[ind])
        final.append(pos)
    return pd.Series([ranks,final])


#team_sum[["ranks","Position"]]= team_sum["index"].apply(lambda x: rank_array(x))
#team_sum["new_ranks"] = team_sum["ranks"].apply(lambda row: rank_keep(row[["ranks"]],row["index"]),axis = 1)
team_sum[["ranks","Position"]] = team_sum.apply(lambda row: rank_array(row["index"]), axis = 1)


team_sum.head(2)
```
