## Create league and season averages for home goals and away goals to use as base for indexing

```python
def season_average(df):
    '''
    Creates rolling season averages for homegoals and awaygoals
    '''
    season_av_df = df.groupby(['league','season','season_gw']).agg({"homegoals":"sum"
                                                                       ,"awaygoals":"sum"
                                                                       ,"fixture":"nunique"}).reset_index()

    season_totals = season_av_df.groupby(['league','season'])["homegoals","awaygoals","fixture"].transform(lambda row: row.rolling(window=36,min_periods=2).sum()).shift(1)

    league_totals = season_av_df.groupby(['league'])["homegoals","awaygoals","fixture"].transform(lambda row: row.rolling(window=36,min_periods=2).sum()).shift(1)

    season_av_df[["TotalHomeGoals","TotalAwayGoals","TotalFixtures"]] = season_totals
    season_av_df["SeasonHG"]=season_av_df["TotalHomeGoals"]/season_av_df["TotalFixtures"]
    season_av_df["SeasonAG"]=season_av_df["TotalAwayGoals"]/season_av_df["TotalFixtures"]
    season_av_df = season_av_df[["league","season","season_gw","SeasonHG","SeasonAG"]]
    return season_av_df
```

## Join team specific stats on to a single line to be able to create fixture predictions

```python
def home_vars_prep(teams_df,join_list,keep_list):
    new_df = teams_df[join_list + keep_list]
    new_df = new_df.rename(columns = {col: col + '_Home' for col in keep_list})
    return new_df

def away_vars_prep(teams_df,join_list,keep_list):
    new_df = teams_df[join_list + keep_list]
    new_df = new_df.rename(columns = {col: col + '_Away' for col in keep_list})
    return new_df

join_list = ['league','season','fixture','season_gw','AverageMethod']
keep_list = ["Team","xG_av","xGA_av","AttStr","DefStr"]

home_df = home_vars_prep(team_stats_df,join_list,keep_list)
away_df = away_vars_prep(team_stats_df,join_list,keep_list)

season_av_df = season_average(results_df)
#home_df.head()
#fixture_vars = pd.merge(home_df,away_df, on = join_list)
#.query("HomeGame==1").drop(columns="HomeGame")

single_line = results_df[["league","season","fixture","season_gw","hometeam","awayteam","homegoals","awaygoals"]].copy()

def final_result(homegoals,awaygoals):
    if homegoals>awaygoals:
        return "H"
    elif homegoals==awaygoals:
        return "D"
    elif homegoals <awaygoals:
        return "A"
    else:
        return np.nan

single_line["Result"] = single_line.apply(lambda row: final_result(row["homegoals"],row["awaygoals"]), axis=1)



single_line = single_line.rename(columns={"hometeam":"Team_Home"
                                         , "awayteam":"Team_Away"})

single_line = pd.merge(single_line,home_df, on=['league', 'season', 'fixture', 'season_gw', 'Team_Home'])
single_line = pd.merge(single_line,away_df, on=['league', 'season', 'fixture', 'season_gw', 'Team_Away',"AverageMethod"])
single_line = pd.merge(single_line,season_av_df, on=['league', 'season', 'season_gw'])


single_line.head()

    
```

## Use indexing to create predicted home and away goals

With our team stats now restructured for every fixture, we will use our attacking and defensive strengths to predict results.

For example Team A has attacking strength of 1.5 and a defensive strength of 0.5. <br>
Team B has an attacking strength of 0.5 and a defensive strenght of 1.5 <br>
Team A is good at attacking and good at defending. Team B is not.

We will multiply Team A's attacking strength against Team B's defensive strength (1.5*1.5) giving us a home goals multiplier of 2.25. Essentially this is saying we're expecting 2.25 more home goals than the average game. The average number of homegoals per game is (hypothetically) 1.5. 2.25 * 1.5 = 3.25. So we're expecting 3.25 home goals in this game.

We can do the same for away goals. Team A has a defensive strength of 0.5 and Team B has an attacking strength of 0.5. 0.5*0.5 = 0.25. 0.25*1.2 (the hypothetical average number of away goals per game) = 0.3

Our predicted score for this hypothetical game is 3.25 - 0.3. <br>
This is based on an above average attacking team playing at home against a below average defensive team, and an above average defensive team playing at home against a below average attacking team.

These values will then be used in Poisson distribution calculations to create probabilities for scores and results.

```python
def Predicted_HG(HAtt,ADef,LgHG):
    Pred_HG = HAtt*ADef*LgHG
    return Pred_HG

def Predicted_AG(HDef,AAtt,LgAG):
    Pred_AG = HDef*AAtt*LgAG
    return Pred_AG

single_line["Pred_HG"] = single_line.apply(lambda row: Predicted_HG(
                                            row["AttStr_Home"],row["DefStr_Away"],row["SeasonHG"]), axis =1)
single_line["Pred_AG"] = single_line.apply(lambda row: Predicted_AG(
                                            row["DefStr_Home"],row["AttStr_Away"],row["SeasonAG"]), axis =1)

single_line[single_line["season_gw"]==3].head(20)
```

## Poisson Distribution to calculate the probability of goals scored

By creating a predicted number of home goals (eg 1.2). We can use poisson distribution to calculate the probability
of 0 home goals,1 home goal, 2 home goals etc.

We can then do the same thing for away goals.

By multplying the two arrays together we can create a score matrix to get the probablity any result. <br>
From here we can calculate the most likely score, as well as probabilities of home,draw and away win results.

```python
#apply poisson distribution to predicted home and away goals. Depending on the number of fixtures you have this can take a minute.

single_line["Prob_HG"] = single_line["Pred_HG"].apply(lambda x: poisson(x).pmf(range(0,10)))
single_line["Prob_AG"] = single_line["Pred_AG"].apply(lambda x: poisson(x).pmf(range(0,10)))
single_line["score"] = single_line.apply(lambda row: pd.DataFrame(np.outer(row["Prob_HG"],row["Prob_AG"])*100), axis =1)
single_line[single_line['season_gw']==3].head()
```

## Create dataframes in the shape of diagonals to be able to extract home, away and draw groupings of scores

By creating dataframes of 1s and 0s for all results that equate to a home win, a draw or an away win, we can multiply the two dataframes together to keep only the probabilities for home wins, which can be summed up to create a home win probability

```python
def blank_mat():
    arr = []
    for num in range(0,10):
        arr.append(0)
    row_vec = np.array(arr)
    row_vec.shape = (1,10)

    arr2 = []
    for num in range(0,10):
        arr2.append(0)

    arr2 = np.array(arr2)
    col_vec = arr2.reshape(-1, 1)
    df_blank = pd.DataFrame(row_vec*col_vec)
    return df_blank

df_blank = blank_mat()
    
def home_mat():
    home_mat = df_blank.copy()
    for i in range(0,10):
        for j in range(0,10):
            if i>j:
                home_mat.loc[i,j] = 1
    return home_mat

home_test = home_mat()
def away_mat():
    away_mat = df_blank.copy()
    for i in range(0,10):
        for j in range(0,10):
            if i<j:
                away_mat.loc[i,j] = 1
    return away_mat
def draw_mat():
    draw_mat = df_blank.copy()
    for i in range(0,10):
        for j in range(0,10):
            if i==j:
                draw_mat.loc[i,j] = 1
    return draw_mat
def wrong_mat():
    wrong_mat = df_blank.copy()
    for i in range(0,10):
        for j in range(0,10):
                wrong_mat.loc[i,j] = 1
    return wrong_mat
def gd_points():
    gd_points_mat = df_blank.copy()
    for i in range(0,10):
        for j in range(0,10):
            if i == j:
                gd_points_mat.loc[i,j] = 20
            else:
                gd_points_mat.loc[i,j] = 15
    return gd_points_mat
    

home_df = home_mat()
away_df = away_mat()
draw_df = draw_mat()
wrong_df = wrong_mat()
gd_points_mat = gd_points()
```

## From the score  matrix to be able to calculate win and score probabilities

We can calculate the most likely score and the most likely result. <br>
We will also calculate the most likely score within the most likely result, eg. if the most likely score is 1-1, but the most likely result is a home win, we can calculate what is more likely, 1-0 or 2-1.

Depending on the number of fixtures you are using the following code can take a few minutes to run

```python

single_line["HomeProb"] = single_line["score"].apply(lambda x: (x*home_df).values.sum())
single_line["DrawProb"] = single_line["score"].apply(lambda x: (x*draw_df).values.sum())
single_line["AwayProb"] = single_line["score"].apply(lambda x: (x*away_df).values.sum())

def most_likely_score(score):
    mx = score.values.max()
    k = score.values

    if np.isnan(mx)==False:
        hg = np.where(k==mx)[0][0]
        ag = np.where(k==mx)[1][0]
        score = str(hg) + " . " + str(ag)
        score_prob = mx
        #return np.array(hg,ag)
        return [score,score_prob]


single_line[["PredScore","PredScoreProb"]] = single_line["score"].apply(lambda x: pd.Series(most_likely_score(x)))

def most_likely_result(Home,Draw,Away):
    mx = np.array([Home,Draw,Away]).max()
    #print(mx)
    if np.isnan(mx) ==False:
        if Home == mx:
            result =  "H"
        elif Draw == mx:
            result = "D"
        else:
            result =  "A"
        return [result,mx]


single_line[["PredResult","PredResultProb"]] = single_line.apply(lambda row: 
                                                 pd.Series(most_likely_result(row["HomeProb"]
                                                                              ,row["DrawProb"]
                                                                              ,row["AwayProb"])),axis =1)

def most_likely_result_score(score,result):
    if result == "H":
        k = (score*home_df).values
    elif result == "D":
        k = (score*draw_df).values
    else:
        k=(score*away_df).values
    
    mx = k.max()
    if np.isnan(mx) ==False:
        hg = np.where(k==mx)[0][0]
        ag = np.where(k==mx)[1][0]
        output_score = str(hg) + " . " + str(ag)
        output_prob = mx
        #return np.array(hg,ag)
        return [output_score,output_prob]

single_line[["PredResultScore","PredResultScoreProb"]]= single_line.apply(lambda row: 
                                                                          pd.Series(most_likely_result_score(row["score"]
                                                                                                             ,row["PredResult"]))
                                                                          ,axis = 1)


def final_result_prob(row):
    if row["Result"]=="H":
        return row["HomeProb"]
    elif row["Result"] =="D":
        return row['DrawProb']
    else:
        return row["AwayProb"]
    
single_line["FinalResultProb"] = single_line.apply(lambda row: final_result_prob(row),axis=1)

single_line[single_line["season_gw"]>2].head(25)
```

```python
def create_final_score(home,away):
    if np.isnan(home)==False:
        h = str(int(home))
        a = str(int(away))
        score_out = h + " . " + a
        return score_out 
        

single_line['FinalScore'] = single_line.apply(lambda row: create_final_score(row["homegoals"],row["awaygoals"]), axis = 1)

single_line["PredCorrect"] = np.where(single_line["Result"]==single_line["PredResult"],1,0)
single_line["PredResultScoreCorrect"] = np.where(single_line["FinalScore"] == single_line["PredResultScore"],1,0)
single_line["PredScoreCorrect"] = np.where(single_line["FinalScore"] == single_line["PredScore"],1,0)
```
